```python
import pandas as pd 
qt = pd.read_csv('Quartly.csv')


print(qt)
```

```python
q26 = pd.read_csv('Q2026.csv')

print(q26)
```

# Definitions
1. Logic = is the brain of the processor. 
2. CowOs = Chips-on-Wafer-Substract is the "super-glue" process used to assemble multiple complex chips together into a single, high-powered system. It is the foundational hardware technology enabling modern artificial intelligence. for example (H100 made by Nvidea)
3. HBM = High-Bandwidth Memory is a specialized, ultra-fast computer memory interface used in high-end graphics cards, server processors, and artificial intelligence accelerators. for example (DDR5)
4. Auxiliary in Hardware = refers to secondary, supplementary, or backup hardware components. These parts do not perform the primary computation, but the system cannot run safely, efficiently, or reliably without them.

```python
T = qt["Logic cost (USD) (median)"].sum()
o = qt["CoWoS cost (USD) (median)"].sum()
t = qt["HBM cost (USD) (median)"].sum()
a = qt["Auxiliary cost (USD) (median)"].sum()

l = "Total cost of Components from 2024-2025 "

total = T + o + t + a 

print(l)
print(f"Logic cost (USD): {T}")
print(f"CoWoS cost (USD): {o}")
print(f"HBM cost (USD): {t}")
print(f"Auxiliary cost (USD): {a}")
print(f"Total Cost (USD): {total}")
```

```python
N = q26["Logic cost (USD) (median)"].sum()
e = q26["CoWoS cost (USD) (median)"].sum()
w = q26["HBM cost (USD) (median)"].sum()
y = q26["Auxiliary cost (USD) (median)"].sum()

r = "Total cost of Components from 2026 "

New_yr = N + e + w + y

print(r)
print(f"Logic cost (USD): {N}")
print(f"CoWoS cost (USD): {e}")
print(f"HBM cost (USD): {w}")
print(f"Auxiliary cost (USD): {y}")

print(f"Total Cost (USD): {New_yr}")
```

```python
import matplotlib.pyplot as plt

labels = ['2024-2025 Total', '2026 Total']
sizes = [total, New_yr]
colors = ['#1f77b4', '#ff7f0e']
explode = (0.05, 0)  # Explode the 2024-2025 slice slightly

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Comparison of Total AI Hardware Costs (2024-2025 vs. 2026)')
plt.tight_layout()
```

# OVERVIEW OF THE DATA ANALYSIS

The anaylsis shows that the cost of hardware from 2024-2026 cost trend has gone down. However, this analysis Regarding the Manufacturing cost of Components that has been lowered across the time period due to factory optimizations such as Self tuning process control and Predictive Maintanance. However, the retail cost has increased since value of AI has rised and the market is willing to pay more. RetaIL pricing reflects the risk, maintenance, and comprehensive system engineering of deploying massive data centers, rather than just the clean component totals found in a raw hardware ledger.

To conclude, Manufacturing data is a reflection of the technology to build hardware is evolving and getting cheaper to replicate. However, the retail price considers market desperation and the cost to hire thousands software developers to build software stacks such as NVIDEA company developeed  Cuda which the developers use to write Ai code and AMD uses ROCm. Moreover, the cooling systems and other maintance of these hardware effect the retail price.
